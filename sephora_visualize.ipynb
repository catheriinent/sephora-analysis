```python
#install needed libraries 
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud as wc

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
data = pd.read_csv('/Users/catherine/python/product_info.csv')
```

```python
#to distinguish products, include the brand name before the product name
data['full_name'] = data['brand_name'] + ' ' + data['product_name']
```

```python
#GRAPH: most "loved"/liked products on Sephora website
most_loved = data.sort_values(by='loves_count', ascending=False)
```

```python
most_loved['full_name'] = most_loved['brand_name'] + ' ' + most_loved['product_name']
```

```python

sns.barplot(data=most_loved, x="loves_count", y='full_name')

#adding labels and title
plt.xlabel('Loves Count')
plt.ylabel('Product Name')
plt.title('Most Loved Products')
#displaying the plot
plt.show()
```

Evidently, the most loved product (by a large margin) is Rare Beauty's Soft Pinch Liquid Blush. The blush is extremely pigmented and rightfully judged, however, it can be said that the large gap between the loves count of the blush compared to the other products in the graph can be acreddited to her loyal fanbase she has garnered throughout her career. By "loving" the product, fans are able to support one of their favourite artists in other ways than liking an Instagram post. Gomez' trajectory of a singer/influencer turned beauty mogul is not a unique story, as seen through Rihanna's Fenty Beauty who also appears multiple times on the list. However, that is not to say that every celebrity can find success just through their fanbases without quality products. Kylie Jenner's Kylie Cosmetics has seen a major decline in attraction despite her hugely influential lip kits, as a result of her products not being high in quality despite major backing in her company. 

Out of the 20 products, 7 of the products are lip products. Compared to having only one "main" foundation, many users will purcahse multiple lip products to fit different occasions. Furthermore, lip products have different types, such as gloss, stains, oils, and lipstick. Thus, where there are a very small number of foundations in the top loved products as a result of users mainly sticking to one foundation to always use, users are able to experiment with different lip products and can love multiple.

```python
#GRAPH: most expensive products on Sephora website (converting USD price to CAD)
most_expensive = data.sort_values(by='price_usd', ascending=False)


most_expensive['full_name'] = most_expensive['brand_name'] + ' ' + most_expensive['product_name']
most_expensive['price_cad'] = most_expensive['price_usd'] * 1.39

```

Within the graph, there are not any makeup products. Instead, the most expensive products are skincare, facial devices, and fragrences. A common theme within the products is "anti-aging", as seen through adjectices such as "regenerating" and "toning". Those who are able to purchase products at these price points are usually older in age, thus the incentive to promote "anti-aging". Furthermore, unlike makeup, these "anti-aging" products work to change the bodily chemistry, rather than settle on top of a customer's face, thus the high price point. 

It is important to distinguish between products that have a 5 star rating because of a small amount of 5 star reviews, and products that have a 5 star rating even having thousand reviews. Thus, the following 4 graphs will represent the unweighted and weighted (products with 5000+ reviews) rated products.

```python
#GRAPH: (Unweighted) Highest rated products

highest_rating = data.sort_values(by='rating', ascending=False)
highest_rating['full_name'] = highest_rating['brand_name'] + ' ' + highest_rating['product_name']
highest_rating = highest_rating.head(20)
sns.barplot(data=highest_rating, x="rating", y='full_name')

plt.xlabel('Rating')
plt.ylabel('Product Name')
plt.title('Highest Rated Products')


plt.show()
```

```python
#GRAPH: (Weighted) Highest rated products
data['num_reviews'] = pd.to_numeric(data['reviews'], errors='coerce')
data = data.dropna(subset=['num_reviews'])
data['num_reviews'] = data['num_reviews'].astype(int)

weighted_reviews = data[data['num_reviews'] >= 5000]
weighted_reviews = weighted_reviews.sort_values(by='rating', ascending=False)
weighted_reviews = weighted_reviews.head(20)
```

```python
sns.barplot(data=weighted_reviews, x="rating", y='full_name')

plt.xlabel('Rating')
plt.ylabel('Product Name')
plt.title('Highest Rated Products (with +5k reviews)')
plt.xlim(4,5)

plt.show()
```

None of the products from the unweighted graph appear on the weighted graph. The weighted graph is a better representation of the actual highest rated products from Sephora.

Anastasia Beverly Hills (ABH) occupies the top 3 spots on the weighted graph. During the rise of influencers promoting makeup, ABH' products were often mentioned due to their strong pigment and overall great quality. Many popular makeup influencers and celebrities such as Kylie Jenner, opted for strong and bold makeup looks, which needed strong pigments such as ABH products. The ABH pomade is a staple in the community.

The NARS blush also appears at a high position in both the weighted rating graph and the "Most Loved" graph. Along with Rare Beauty's blush's prominence in the "Most Loved" graph, that a well-crafted blush is highly valued within the community and can become a staple of a brand. 

Despite lip products appearing constantly on the "Most Loved" graph, they rarely appear on the "Highest Rated Products" graph. Through this observation, one may be able to conclude that customers can be more forgiving towards a lower quality lip product than they are with eye products, which are prominent on the "Highest Rated Product" graph.

```python
#GRAPH: (Unweighted) Lowest rated product
lowest_rating = data.sort_values(by='rating')
lowest_rating['full_name'] = lowest_rating['brand_name'] + ' ' + lowest_rating['product_name']
lowest_rating = lowest_rating.head(20)
sns.barplot(data=lowest_rating, x="rating", y='full_name')

plt.xlabel('Rating')
plt.ylabel('Product Name')
plt.title('Lowest Rated Products')


plt.show()
```

Similar to the unweighted and weighted versions of the "Highest Rated Products" graph, the unweighted and weighted versions of the "Lowest Rated Products" graph do not share any common products. However, the range for the ratings is larger for the lowest rated products than it is for the highest rated product (2.9-3.6 against 4.5-5). 

Although the products on the weighted graph are the "Lowest Rated Products", the lowest rating hovers around 3, thus it is still a decent product, just not stellar. Similarly to the "Highest Rated Products", many of the products on the weighted "Lowest Rated Products" graph are eye products. People are less forgiving towards a bad eye product than any other product.

```python
#word cloud for most commonly used words in reviews
rdata = pd.read_csv('/Users/catherine/python/conjoined.csv')
```

```python
rdata['review_text'] = rdata['review_text'].astype(str).fillna('')
```

```python
text = " ".join(review for review in rdata['review_text'])
wordcloud = wc(width=800, height=400, background_color='white').generate(text)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Remove axes
plt.show()
```

I will avoid analysis of words such as: "product", "using", "skin", and "used", which are words that have no aid in letting the viewer know what consumers value most in a product.

When paying attention to customer critique, this word cloud distinguishes what attributes of makeup that are most important to them when writing a review. 

"Sensitive skin": Sensitive skin is not uncommon, and for many people, their sensitive skin can cause them to turn away from experimenting with makeup due to the fear of breaking out. When creating products, companies should pay attention to incorporating clean products that do not upset sensitive skin (such as staying away from scented ingredients). 

"Eye cream": As eye creams do not appear within the top 20 most reviewed products, the prominence of the phrase "eye cream" in the word cloud can possibly mean that many users are reviewing products with respect to how the product reacts to their eye cream. Does the foundation/concealer lay nicely on top? Or does it pill and seperate? Eye creams are extremely popular inclusions within skincare routines, thus it is beneficial for companies to keep in mind that their product should not only work nicely with other products in their brand, but with others. 

"Skin feeling": Although a product may look excellent in photos and in real life, companies should concern themselves with how the user will feel after they apply the product. Will their skin feel dried out? Oily? Offputting after-effects can turn customers away from purchasing products.
